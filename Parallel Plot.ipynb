```python
df = df[(df != 0).all(1)]
```

```python
df18 = df.loc[df['year']==2018]
df18.head()
```

```python
df18['age'] = df18['age'].astype('category')
df18['cat_codes'] = df18['age'].cat.codes
df18.head()
```

```python
# Create dimensions
type_dim = go.parcats.Dimension(
    values=df18.facility_type_a,
    categoryorder='category ascending',
    label="Facility Type"
)

sex_dim = go.parcats.Dimension(
    values=df18.sex,
    label="Sex"
)

age_dim = go.parcats.Dimension(
  values=df18.age,
  #counts=df_melt.value,
  label="Age",
  #categoryarray=['Shopping','Accommodation','FnB','Other'],
  #ticktext=['perished', 'survived'],
)

# Create parcats trace
#'Shopping','Accommodation','FnB','Other'
color = df18.cat_codes;
#colorscale = [[0, 'rgb(136, 204, 238)'], [1, 'rgb(204, 102, 119)'],[2, 'rgb(221, 204, 119)'],[3, 'rgb(68, 170, 153)']];

data = [
    go.Parcats(
        dimensions=[type_dim, sex_dim, age_dim],
        counts=df18.rate,
        line={'color': color,
              'colorscale': 'earth'},
        hoveron='category',
        hoverinfo='count+probability',
        labelfont={'size': 18, 'family': 'Times','color':'saddlebrown'},
        tickfont={'size': 10, 'family': 'Times'},
        arrangement='freeform'
    )
]

layout=go.Layout(
        title="Admission Rate Per 1,000 Resident Population by Age and Sex",
        height=500,
        )


# Display figure
fig = go.Figure(data=data, layout =layout)

#iplot(data)
```
